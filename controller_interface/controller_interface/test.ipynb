```python
import rclpy
from rclpy.node import Node
from mcp_ep_msgs.msg import MujocoApplyForceIndex
import time
```

```python
class ControllerInterfaceNode(Node):
    
    def __init__(self):
        super().__init__('controller_interface_node')
        self.publisher_ = self.create_publisher(MujocoApplyForceIndex, 'mujoco_apply_force_index', 10)
        
    def mujoco_apply_force_index_pub(self, face_indices, link_indicies, forces_norm):
        msg = MujocoApplyForceIndex()
        msg.face_index = face_indices
        msg.link_index = link_indicies
        msg.force_norm = forces_norm
        self.publisher_.publish(msg)
        
    def mujoco_apply_force_index_clear(self):
        msg = MujocoApplyForceIndex()
        msg.face_index = []
        msg.link_index = []
        msg.force_norm = []
        self.publisher_.publish(msg)
```

```python
rclpy.init(args=None)
node = ControllerInterfaceNode()
```

```python
link_indicies = [4]
face_indicies = [69]
forces_norm = [15.0]
node.mujoco_apply_force_index_pub(face_indicies, link_indicies, forces_norm)
```

```python
for i in range(1000):
    node.mujoco_apply_force_index_clear()
    time.sleep(0.5)
    face_indicies[0] = i
    print(face_indicies[0])
    node.mujoco_apply_force_index_pub(face_indicies, link_indicies, forces_norm)
    time.sleep(1)
```
